```python
import matplotlib.pyplot as plt

import pandas as pd

data = pd.read_csv('data.csv')

data.head()
```

```python
# importing the module
import matplotlib.pyplot as plt

# setting the size
plt.figure(figsize=(15,8))

# plotting the graph
plt.plot([i for i in range(len(data))], data['water level'])
plt.show()
```

```python
# importing the module
import matplotlib.pyplot as plt

# setting the size
plt.figure(figsize=(15,8))

# plotting the graph
plt.plot([i for i in range(len(data))], data['error'])
plt.show()
```
