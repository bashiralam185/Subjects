###### c) Let's try to make a graph reading from a file:
<p> As you can see in the â€œsizesâ€ is a python list of numbers. And â€labelsâ€ is a series of strings.
Now imagine that we have a text file containing the labels on the first line and the sizes on
the second line. Strings and numbers are all separated by a coma with no space.
Write the program to read the data from the file and create the graph.</p>

```python
with open('file.txt') as f:
    lines = f.readlines()
```

```python
# creating lists for the labels and sizes
label =[]
size =[]

# reading the lines, checking is the word is number or label
with open("file.txt","r") as f:
    for line in f:
        for word in line.split(","):
            try:
                size.append(int(word))
            except:
                label.append(word)

                
explode = (0, 0.1, 0, 0) #get that slice out
fig1,ax1=plt.subplots()
ax1.pie(size,explode = explode,labels=label,autopct='%1.1f%%',startangle=90)
ax1.axis('equal')
plt.show()
```

```python
explode = (0, 0.1, 0, 0) #get that slice out
fig1,ax1=plt.subplots()
ax1.pie(size,explode = explode,labels=label,autopct='%1.1f%%',startangle=90,wedgeprops={'width':0.3})
ax1.axis('equal')
plt.show()
```

```python
import PyPDF2
pdf = PyPDF2.PdfFileReader(open("practical 1.pdf", "rb"))
for page in pdf.pages:
    file1 = page.extractText()
    with open('main_file.txt', 'a') as f:
        f.write(file1)

# list of all necessary words
all_words = ["of", 'the', 'to', 'that']
dic = {}

# opening the file and using for loop to check and count the words
with open("main_file.txt", 'r') as f:
    for line in f:
        words = line.split()
        for i in words:
            if i in all_words:    
                if i in dic:
                    dic[i] = dic[i] + 1
                else:
                    dic[i] = 1

dic
```

```python
fig1,ax1=plt.subplots()
ax1.pie(dic.values(),labels=dic.keys(),autopct='%1.1f%%',startangle=90,wedgeprops={'width':0.3})
ax1.axis('equal')
plt.show()
```
