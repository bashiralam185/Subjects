```python
#subset of the dashfired safety project
import pandas as pd
import datetime as dt
from datetime import timedelta
import matplotlib.pyplot as plt



df=pd.read_excel('dashboard.xlsx')
#1
totac=df.groupby('Date')['Outcome'].count()
totsuc=df[df['Outcome']=='Success'].groupby('Date')['Outcome'].count()
totnsuc=df[df['Outcome']=='Failure'].groupby('Date')['Outcome'].count()
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
axes.plot(totac.index, totac, 'r',label='Total')
axes.plot(totsuc.index, totsuc, 'b',label='Success')
axes.plot(totnsuc.index, totnsuc, 'g',label='Failure')

dfsd=totac[totsuc.index]
axes.plot(totsuc.index, totsuc*100/dfsd, 'orange',label='ratio%')

axes.set_xlabel('Dates')
axes.set_ylabel('numbers')
axes.set_title('Action repport')
axes.legend(loc=2)
# plt.show()
```

```python
import matplotlib.gridspec as gridspec

def create_figure():
    with plt.style.context(("seaborn","ggplot")):
            fig = plt.figure(constrained_layout=True, figsize=(10,8))
            specs = gridspec.GridSpec(ncols=2, nrows=2, figure=fig) ## Declaring 2x2 figure.

            ax1 = fig.add_subplot(specs[0, :]) ## First Row
            ax2 = fig.add_subplot(specs[1, 0]) ## Second Row First Column

            
            
            
            
            ax1.plot(totac.index, totac, 'r',label='Total')
            ax1.plot(totsuc.index, totsuc, 'b',label='Success')
            ax1.plot(totnsuc.index, totnsuc, 'g',label='Failure')

            dfsd=totac[totsuc.index]
            ax1.plot(totsuc.index, totsuc*100/dfsd, 'orange',label='ratio%')

            ax1.set_xlabel('Dates')
            ax1.set_ylabel('numbers')
            ax1.set_title('Action repport')
            ax1.legend(loc=2)
    
    
    
    

    
            ax2.plot(totac.index, totac,'r',ls='None',marker='o' ,label='Total')
            ax2.plot(totsuc.index, totsuc, 'b',ls='None',marker='+',label='Success')
            ax2.plot(totnsuc.index, totnsuc, 'g',ls='None',marker='s',label='Failure')
            dfsd=totac[totsuc.index]
            ax2.plot(totsuc.index, totsuc*100/dfsd,'orange',ls='None', marker='1',label='ratio%')
            ax2.set_xlabel('Dates')
            ax2.set_ylabel('numbers')
            ax2.set_title('Action repport')
            ax2.legend(loc=2)
    


    
    
            a='Failure','Success'
            sizes=[df[df['Outcome']=='Success'].shape[0],df[df['Outcome']=='Failure'].shape[0]]
            explode = (0, 0.1)  #get that slice out
            fig1,ax3=plt.subplots()
            ax3.pie(sizes,explode = explode,labels=a,autopct='%1.1f%%',shadow=True,startangle=90)
            ax3.axis('equal')
            plt.title('succes vs failure')
            
            
            
            
            
            
            
            
            
            
            
#             totac=df.groupby('State')['Outcome'].count()
#             totsuc=df[df['Outcome']=='Success'].groupby('State')['Outcome'].count()
            fig, ax = plt.subplots(figsize=(12,6))
            size = 0.3

            ax.pie(totac.values.flatten(), radius=1,
                   labels=totac.index,
                   autopct='%1.1f%%',
                   wedgeprops=dict(width=size, edgecolor='w'))

            ax.set(aspect="equal", title='Pie plot with `ax.pie`')

            plt.close(fig)
            return fig
```

```python
create_figure()
```

```python
import panel as pn
pn.extension()

kw = dict()

dash1 = pn.interact(create_figure, **kw)
dash1
```
